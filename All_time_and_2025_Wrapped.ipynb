```python
import pandas as pd
import json
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ruta = '../data/*.json'
archivos = glob.glob(ruta)
dfs = []
for archivo in archivos:
    with open (archivo, 'r',encoding = 'utf-8') as f:
        data = json.load(f)
        df = pd.DataFrame(data)
        dfs.append(df)
spoti = pd.concat(dfs, ignore_index = True)
spoti
```

```python
spoti.info()
```

```python
spoti[spoti.duplicated(keep=False)] #Comprobar duplicados
```

```python
spoti = spoti.drop_duplicates() #Eliminar duplicados
spoti.info() #Observamos que el número de 'entries' ha disminuido
```

<h1>ALL-TIME WRAPPED</h1>

<h2>TOP 10 ARTISTAS</h2>
<strong>Junto los minutos totales de escucha</strong>

```python
spoti['minutes'] = spoti['ms_played']/60000
artistas_top = (spoti.groupby('master_metadata_album_artist_name')['minutes'].sum().sort_values(ascending=False).head(10))
df_artistas = pd.DataFrame(artistas_top)
df_artistas
```

<strong>Gráficas con los minutos de escucha por año</strong>

```python
spoti['ts'] = pd.to_datetime(spoti['ts'])
spoti['year'] = spoti['ts'].dt.year
```

```python
for artist in artistas_top.index:
    datos = (spoti[spoti['master_metadata_album_artist_name'] == artist].groupby('year')['minutes'].sum() )

    datos.plot(kind='bar', figsize=(7,4))
    plt.title(f"Minutos por año — {artist}")
    plt.xlabel("Año")
    plt.ylabel("Minutos escuchados")
    plt.tight_layout()
    plt.show()
```

<H2>TOP 10 CANCIONES</H2>
<strong>Junto los minutos totales de escucha, el artista y su correspondiente álbum</strong>

```python
canciones_top = (spoti.groupby(['master_metadata_track_name','master_metadata_album_artist_name','master_metadata_album_album_name'])['minutes'].sum().sort_values(ascending=False).head(10))
df_canciones = pd.DataFrame(canciones_top)
df_canciones
```

<strong>Gráficas con los minutos de escucha por año</strong>

```python
top_songs = df_canciones.index.get_level_values('master_metadata_track_name').tolist()

for song in top_songs:
    datos = (spoti[spoti['master_metadata_track_name'] == song].groupby('year')['minutes'].sum().round())

    datos.plot(kind='bar', figsize=(7,4))
    plt.title(f"Minutos por año — {song}")
    plt.xlabel("Año")
    plt.ylabel("Minutos escuchados")
    plt.tight_layout()
    plt.show()
```

<H2>TOP 10 ÁLBUMES</H2>
<strong>Junto los minutos totales de escucha y el/los artista/s</strong>

```python
albumes_top = (spoti.groupby(['master_metadata_album_album_name','master_metadata_album_artist_name'])['minutes'].sum().round().sort_values(ascending=False).head(10))
df_albumes = pd.DataFrame(albumes_top)
df_albumes
```

<strong>Gráficas con los minutos de escucha por año</strong>

```python
top_albums = df_albumes.index.get_level_values('master_metadata_album_album_name').tolist()

for album in top_albums:
    datos = (spoti[spoti['master_metadata_album_album_name'] == album].groupby('year')['minutes'].sum().round())

    datos.plot(kind='bar', figsize=(7,4))
    plt.title(f"Minutos por año — {album}")
    plt.xlabel("Año")
    plt.ylabel("Minutos escuchados")
    plt.tight_layout()
    plt.show()
```

<h2>ESCUCHAS POR HORA DEL DÍA</h2>

```python
spoti['hour'] = spoti['ts'].dt.hour
min_horas_dia_años = spoti.groupby('hour')['minutes'].sum().round()
#Gráfica
min_horas_dia_años.plot(kind='bar', figsize=(8,4))
plt.title("Minutos escuchados por horas del día")
plt.xlabel("Hora del día")
plt.ylabel("Minutos")
plt.tight_layout()
plt.show()
```

<h2>DÍA CON MÁS ESCUCHAS DE CADA AÑO</h2>

```python
# DataFrame auxiliar
tmp = spoti.copy()

tmp['year'] = tmp['ts'].dt.year
tmp['date_only'] = tmp['ts'].dt.date
mins_by_day = (tmp.groupby(['year', 'date_only'])['minutes'].sum().round().reset_index())

top_day_each_year = mins_by_day.loc[mins_by_day.groupby('year')['minutes'].idxmax()]

top_day_each_year
```
